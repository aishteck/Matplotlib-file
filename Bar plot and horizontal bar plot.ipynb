## Bar plot and horizontal bar plot

A bar plot helps us understand the distribution of categorical data.

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Bar plot 

Here, we are going to create a dictionary first.

```python
student = {"Ali": 87 , "Aisha" : 67, "Anaya" : 27}
student
```

```python
names = list(student.keys())
names
```

```python
values = list (student.values())
values
```

```python
plt.bar(names,values)
plt.show()
```

**Adding title and labels**

```python
plt.bar(names, values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
