# First plot

## Line plot

we are going to import libraries, numpy and matplotlip first

```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
x = np.arange (1,11)
x
```

```python
y =2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

## Title x-lebal and  y-lebal

```python
plt.plot(x,y)
plt.title("First Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## Changing line Plot
**Changing line Aesthetics**

```python
plt.plot(x,y, color='b', linestyle=':', linewidth=2)
plt.title("First Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# line Plot
**Adding two line in the same plot**

```python
x = np.arange (1,11)
y1 = 2*x
y2 = 3*x
x,y1,y2
```

```python
plt.plot(x,y1, color='y', linestyle= ':', linewidth=5)
plt.plot(x,y2, color='g' , linestyle='-', linewidth=5)
plt.title("two line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y1, color='y', linestyle= ':', linewidth=5)
plt.plot(x,y2, color='g' , linestyle='-', linewidth=5)
plt.title("two line Plot without gri")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
