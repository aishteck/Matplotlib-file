## Line Plot 
**Adding Sub-plot**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
y = 2*x
z = 3*x
x,y,z
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='g', linestyle=':', linewidth=5)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(y,z, color='gray', linestyle= '-' , linewidth=5)
plt.show()
```

```python
plt.subplot(1,2,1) # here is 1 row and two column and 1 index
plt.plot(x,y, color='g', linestyle=':', linewidth=5)

plt.subplot(1,2,2) # 1 row , two column and two index
plt.plot(y,z, color='gray', linestyle= '-' , linewidth=5)

plt.show()
```

```python
 plt.subplot(1,2,1) # here is 1 row and two column and 1 index
plt.plot(x,y, color='g', linestyle=':', linewidth=5)
plt.grid(True)

plt.subplot(1,2,2) # 1 row , two column and two index
plt.plot(y,z, color='gray', linestyle= '-' , linewidth=5)
plt.grid(True)

plt.show()
```

**Two rows and one column**

```python
 plt.subplot(2,1,1) # here is 2 row and 1 column and 1 index
plt.plot(x,y, color='y', linestyle=':', linewidth=2)
plt.grid(True)

plt.subplot(2,1,2) # 2 row , 1 column and two index
plt.plot(y,z, color='g', linestyle= '-' , linewidth=2)
plt.grid(True)

plt.show()
```
